```python
# Import libraries
import pickle
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import load_model
from PIL import Image
```

```python
# membuat variabel baru untuk model
model = load_model('model_transfer.h5')
```

```python
# Mendefinisikan path file gambar
file_path = "C:\Panji\Hacktiv8\FASE_2\GC7\p2-ftds012-hck-g7-panjiputra-r\Dummy_1.jpg"

if os.path.exists(file_path):
    img = tf.keras.utils.load_img(file_path, target_size=(224, 224))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    x = tf.keras.utils.img_to_array(img)
    x = np.expand_dims(x, axis=0)

    # melakukan predict
    classes = model.predict(x)
    result_pred = np.where(classes < 0.7, 0, 1)
    if result_pred == 1:
        print('Brain_tumor')
    else:
        print('Healthy')
else:
    print("File not found:", file_path)
```
